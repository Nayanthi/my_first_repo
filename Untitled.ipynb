## Regression

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import statsmodels.api as sm
import math
%matlplotlib inline
```

```python
data=datasets.load_boston()
df=pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
data.keys()
```

```python
df['MEDV']= data.target
df.head(10)
```

```python
model= sm.OLS(df.MEDV, sm.add_constant(df.RM))
p= model.fit().params
```

```python
df.RM
```

```python
p
```

## Scatter plot

```python
ax=df.plot(x="RM", y='MEDV', kind='scatter',title='scatter')
ax.set_title=("sactter")  
x=range(math.floor(min(df.RM)),math.ceil(max(df.RM))+1)
matplotlib_plot=ax.plot(x,p.const+ p.RM* x,color='red')
```

```python
import seaborn as sns
ax1= sns.residplot(x=df.RM, y=df.MEDV)
```

```python
df["age_cat"]=['young' if i<45 else 'old' for i in df.AGE]
```

```python
df.age_cat
```

```python
ax3=sns.lmplot(x='RM', y='MEDV', hue='age_cat',data=df)
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(df,vars=['RM', 'AGE', 'MEDV'], hue="age_cat")
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(sns.kdeplot, lw=3, legend=False);
```
